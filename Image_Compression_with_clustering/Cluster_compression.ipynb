```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import k_means
import numpy as npy
```

```python
img = plt.imread('cat.jpg')
img_size = img.shape
print(img.shape)
```

### Reshape image to to 2 dimensional array

```python
img2D = img.reshape(img_size[0] * img_size[1], img_size[2])
print(img2D)
```

### use centroid for each color

```python
from sklearn.cluster import KMeans

numcluster = 6
kmeans = KMeans(n_clusters=numcluster)
theClusters = kmeans.fit_predict(img2D)
x_compressed = kmeans.cluster_centers_[kmeans.labels_]
x_compressed = np.clip(x_compressed.astype('uint8'), 0, 255)
x_compressed = x_compressed.reshape(img_size[0], img_size[1], img_size[2])
```

### plot original and compressed image

```python
fig, ax = plt.subplots(1,2,figsize = (12,8))
ax[0].imshow(img)
ax[0].set_title('Original image')
ax[1].imshow(x_compressed)
ax[1].set_title('Compressed image ' + str(numcluster) + ' Colors')
for ax in fig.axes:
    ax.axis('off')
plt.tight_layout()
```
