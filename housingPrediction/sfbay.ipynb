# Deep Learning

```python
#Clean Data
#Scrub the data, remove entries that are missing
filterData = rawData.dropna(subset=['total_rooms'])
filterData = filterData.dropna(subset='latitude')
filterData = filterData.dropna(subset='longitude')
filterData = filterData.dropna(subset='housing_median_age')
filterData = filterData.dropna(subset='total_rooms')
filterData = filterData.dropna(subset='total_bedrooms')
filterData = filterData.dropna(subset='population')
filterData = filterData.dropna(subset='households')
filterData = filterData.dropna(subset='median_income')
filterData = filterData.dropna(subset='median_house_value')
filterData = filterData.dropna(subset='ocean_proximity')
```

```python
filterData.describe()
```

```python
inputs = filterData.drop(columns=['median_house_value', 'latitude', 'longitude', 'ocean_proximity']) #removing house value because it is output, proximity is categorical, and lat/long as a test
```

```python
outputs = filterData['median_house_value']
```

```python
model = keras.Sequential()
```

```python
model.add(keras.layers.Dense(6, activation='relu', input_shape=(6,))) # input layer: we have 6 inputs: age, room, beds, pop, households, income
```

```python
model.add(keras.layers.Dense(5, activation="relu")) # hidden layer
model.add(keras.layers.Dense(3, activation="relu"))# hidden layer
model.add(keras.layers.Dense(2, activation="relu"))# hidden layer
```

```python
model.add(keras.layers.Dense(1))
```

```python
model.compile(optimizer='adam', loss='mean_squared_error')
```

```python
model.fit(inputs,outputs, epochs=400, batch_size=600)
```
